```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([0,1,2,3])
y=np.array([2,4,6,8])
plt.plot(x,y)
plt.show()
```

###  Plotting Without Line


```python
x_axis=np.array([0,5])
y_axis=np.array([3,7])
plt.plot(x_axis, y_axis,'o') 
plt.show()
```

#### Multiple Points

```python
x=np.array([1,2,6,8])
y=np.array([3,8,1,12])
plt.plot(x,y)
plt.show()
```

### Default X-axis


```python
y_axis=np.array([3,8,1,10,5,7])
plt.plot(y_axis)
plt.show()
```

### Markers

```python
y=np.array([3,5,1,9,4])
plt.plot(y,marker='o')
plt.show()
```

### Linestyle

```python
y=np.array([3,5,1,9,4])
plt.plot(y,marker='o',linestyle="dotted")
plt.show()
```

```python
y=np.array([3,5,1,9,4])
plt.plot(y,marker='o',color='r')
plt.show()
```

###  Line Width


```python
y=np.array([3,5,1,9,4])
plt.plot(y,marker='o',color='r',linewidth="12.3")
plt.show()
```

### Multiple Lines


```python
a1=[3,8,1,9,2];
a2=[6,2,7,11];
plt.plot(a1)
plt.plot(a2)
```

```python
x=np.array([10,20,30,40,50])
y=np.array([12,26,13,18,16])
plt.plot(x,y);
plt.xlabel("pulse")
plt.ylabel("Rate")
```

```python
x=np.array([10,20,30,40,50])
y=np.array([12,26,13,18,16])
plt.plot(x,y);
plt.title("Heart Rate")
plt.xlabel("pulse")
plt.ylabel("Rate")
```

```python
x=np.array([10,20,30,40,50])
y=np.array([12,26,13,18,16])
plt.plot(x,y);
plt.title("Heart Rate",loc="Right")
plt.xlabel("pulse")
plt.ylabel("Rate")
```

```python
x=np.array([10,20,30,40,50])
y=np.array([12,26,13,18,16])
plt.plot(x,y);
plt.title("Heart Rate")
plt.xlabel("pulse")
plt.ylabel("Rate")
plt.grid()
plt.show()
```

```python
x=np.array([10,20,30,40,50])
y=np.array([12,26,13,18,16])
plt.plot(x,y);
plt.title("Heart Rate")
plt.xlabel("pulse")
plt.ylabel("Rate")
plt.grid(axis="x")
plt.show()
```

### Matplotlib Subplot


```python
# plot 1:
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)

# plot 2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

### Matplotlib Scatter


```python
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y) 
plt.show()
```

### Colors

```python
# plot-1
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y,color='pink')

# plot-2
x=np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y=np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y,color='b')
plt.show()
```

### Matplotlib Bars


```python
x=np.array(['A','B','C','D'])
y=np.array([3,8,10,12])
plt.bar(x,y)
plt.show()
```

```python
y=np.array([40,25,25,10])
plt.pie(y)
plt.show()
```

```python
y=np.array([40,25,25,10])
mylabels=["A","B","C","D"]
myexplode=[0.2,0,0.1,0]

plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```

```python
y=np.array([40,25,25,10])
mylabels=["A","B","C","D"]
myexplode=[0.2,0,0.1,0]

plt.pie(y,labels=mylabels,explode=myexplode,shadow="True")
plt.show()
```

```python
y=np.array([40,25,25,10])
mylabels=["A","B","C","D"]
plt.pie(y,labels=mylabels)
plt.legend()

plt.show()
```

```python
y=np.array([40,25,25,10])
mylabels=["A","B","C","D"]
plt.pie(y,labels=mylabels)
plt.legend(title="Alphabets")
plt.show()
```
